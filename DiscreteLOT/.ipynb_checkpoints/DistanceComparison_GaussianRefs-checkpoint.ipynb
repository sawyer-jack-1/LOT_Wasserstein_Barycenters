### Example 11 of "Supervised learning of sheared distributions using linearized optimal transport", V. Khurana, H. Kannan,  A. Cloninger, C. Moosm√ºller (2022)
Reference: https://arxiv.org/abs/2201.10590

```python
#Import required packages
import numpy as np
import matplotlib.pyplot as plt
import ot
import ot.plot
import sys
import math
import os
import seaborn as sns
import Gaussians
```

```python
#Sample m points from reference distribution - multivariate Gaussian distribution
n = 28
m = 10000
sigma = 1

#Mean and covariance for template distribution
mu = np.array([(n-1)/2, (n-1)/2])
cov = pow(sigma,2)*np.identity(2)

x = np.linspace(0,n-1,n)
y = np.linspace(0,n-1,n)

unsheared_samples, template = Gaussians.DiscreteGaussian(mu,cov,x,y,n,m)

#Mean and covariance for template distribution
mu = np.array([(n-1)/2, (n-1)/2])
cov = pow(sigma,2)*np.identity(2)

#Mean and covariance for sheared distribution
theta = np.radians(45)
lambda_1 = 1.5
lambda_2 = 0.5
b = 1*np.ones(2)

P = np.asarray([[np.cos(theta), -np.sin(theta)],[np.sin(theta), np.cos(theta)]]) 
Lambda = np.diag([lambda_1, lambda_2])
A = np.matmul(np.transpose(P), np.matmul(Lambda,P))

sheared_mu = np.matmul(A,mu) + b
sheared_cov = np.matmul(A,np.transpose(A))
sheared_samples, sheared_template = Gaussians.DiscreteGaussian(sheared_mu,sheared_cov,x,y,n,m)

#Mean and covariance for reference distribution
lambda_1 = 1.5
lambda_2 = 0.5
d = 3*np.ones(2)
Lambda = np.diag([lambda_1, lambda_2])
C = np.matmul(np.transpose(P), np.matmul(Lambda,P))
reference_mu = np.matmul(C,mu) + d
reference_cov = np.matmul(C,np.transpose(C))
reference_samples, reference = Gaussians.DiscreteGaussian(reference_mu,reference_cov, x,y,n,m)
print(mu)
print(sheared_mu)
print(reference_mu)
```

```python
sns.set_style("white")

plt.rcParams["figure.figsize"] = (15,15)
plt.rcParams["figure.dpi"] = 200

#Plot Grid with the template samples
fig, ax = plt.subplots(3,2)
num_gridPoints = range(0,n)
grid_points=[]
for i in range(0,n):
    for j in range(0,n):
        grid_points.append([x[i], y[j]])

grid_points = np.asarray(grid_points)

for i in range(0,3):
    for j in range(0,2):
        ax[i,j].set_xlim([0,n-1])
        ax[i,j].set_ylim([0,n-1])
        ax[i,j].set_xticks(num_gridPoints)
        ax[i,j].set_yticks(num_gridPoints)
        ax[i,j].xaxis.grid(True)
        ax[i,j].yaxis.grid(True)
        ax[i,j].tick_params(labelbottom=False)    
        ax[i,j].tick_params(labelleft=False) 


ax[0,0].plot(unsheared_samples[:, 0], unsheared_samples[:, 1], 'or',markersize=1)
ax[0,0].set_title('Samples from a template Gaussian distribution')

ax[0,1].scatter(grid_points[:,0],grid_points[:,1],s=250*template)
ax[0,1].set(title="Template samples snapped to a 28 x 28 grid")

ax[1,0].plot(sheared_samples[:, 0], sheared_samples[:, 1], 'or',markersize=1)
ax[1,0].set_title('Samples from sheared Gaussian distribution')

ax[1,1].scatter(grid_points[:,0],grid_points[:,1],s=250*sheared_template)
ax[1,1].set(title="Sheared samples snapped to a 28 x 28 grid")

ax[2,0].plot(reference_samples[:, 0], reference_samples[:, 1], 'or',markersize=1)
ax[2,0].set_title('Samples from candidate reference Gaussian distribution')

ax[2,1].scatter(grid_points[:,0],grid_points[:,1],s=250*reference)
ax[2,1].set(title="Reference samples snapped to a 28 x 28 grid")

#fig.savefig("Images/TestGaussians.png",bbox_inches="tight",dpi=500)
```

```python
#Squared Euclidean Cost matrix between the grid points
Cost = ot.dist(grid_points, grid_points, metric = "sqeuclidean")

LOT_distances = []
Wasserstein_distances = []

for k in range(0,500):
    
    #Mean and covariance for sheared distribution
    theta = np.random.uniform(0,360)
    lambda_1 = np.random.uniform(0.5,2.5)
    lambda_2 = np.random.uniform(0.5,2.5)
    b = np.random.randint(0,5)

    P = np.asarray([[np.cos(theta), -np.sin(theta)],[np.sin(theta), np.cos(theta)]]) 
    Lambda = np.diag([lambda_1, lambda_2])
    A = np.matmul(np.transpose(P), np.matmul(Lambda,P))

    #Shear the template
    sheared_mu = np.matmul(A,mu) + b
    sheared_cov = np.matmul(A,np.transpose(A))
    sheared_samples, sheared_template = Gaussians.DiscreteGaussian(sheared_mu,sheared_cov,x,y,n,m)
    
    #Mean and covariance for reference distribution
    lambda_1 = np.random.uniform(0.5,2.5)
    lambda_2 = np.random.uniform(0.5,2.5)
    d = np.random.randint(0,5)
    Lambda = np.diag([lambda_1, lambda_2])
    C = np.matmul(np.transpose(P), np.matmul(Lambda,P))
    reference_mu = np.matmul(C,mu) + d
    reference_cov = np.matmul(C,np.transpose(C))
    reference_samples, reference = Gaussians.DiscreteGaussian(reference_mu,reference_cov, x,y,n,m)
    
    #Compute optimal transport plan from reference for the template and sheared template
    unsheared_embedTransport  = ot.lp.emd(reference,template,Cost)
    sheared_embedTransport  = ot.lp.emd(reference,sheared_template,Cost)

    #Threshold optimal transport plan to get Monge map
    unsheared_embedding = []
    sheared_embedding = []
    support_1 = []
    support_2 = []
    for i in range(0,unsheared_embedTransport.shape[0]):
        if np.max(unsheared_embedTransport[i]) > 1e-12:
            x_index = math.floor(np.argmax(unsheared_embedTransport[i]) / n)
            y_index = np.argmax(unsheared_embedTransport[i]) % n        
            unsheared_embedding.append([x_index, y_index])
            support_1.append(i)
        if np.amax(sheared_embedTransport[i]) > 1e-12:
            x_index = math.floor(np.argmax(sheared_embedTransport[i]) / n)
            y_index = np.argmax(sheared_embedTransport[i]) % n    
            sheared_embedding.append([x_index, y_index])
            support_2.append(i)

    unsheared_embedding = np.asarray(unsheared_embedding)
    sheared_embedding = np.asarray(sheared_embedding)
    support_1 = np.asarray(support_1)
    support_2 = np.asarray(support_2)
    
    #print(support_1-support_2)
    #print(unsheared_embedding.shape)
    #print(sheared_embedding.shape)
    
    #LOT embedding distance
    LOTdist = 0
    for i in range(0,unsheared_embedding.shape[0]):
        LOTdist = LOTdist + np.linalg.norm(unsheared_embedding[i]-sheared_embedding[i],2)*reference[int(support_2[i])]
    LOT_distances.append(LOTdist)
    
    #Compute wasserstein distance from reference for the template and sheared template
    wassDist = np.sqrt(ot.lp.emd2(template,sheared_template,Cost))
    Wasserstein_distances.append(wassDist)
```

```python
#Scatter plot LOT distances and Wasserstein distances
sns.set_style("whitegrid")

fig, ax = plt.subplots(1,1)
plt.rcParams["figure.figsize"] = (7,5)
plt.rcParams["figure.dpi"] = 200

ax.scatter(LOT_distances, Wasserstein_distances, s=4, c="b")
x = np.linspace(0,np.amax([np.amax(LOT_distances),np.amax(Wasserstein_distances)]))
ax.plot(x,x,"--r",linewidth=2)
ax.set(xlabel="LOT distance", ylabel="Wasserstein distance")
#fig.savefig("Images/Example11.png",bbox_inches="tight",dpi=500)
```
